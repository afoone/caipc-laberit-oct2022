```python
def text_to_number(x):
    if (x == 'four'):
        return 4
    elif (x == 'six'):
        return 6
    elif (x == 'five'):
        return 5
    elif (x == 'eight'):
        return 8
    elif (x == 'two'):
        return 2
    elif (x == 'three'):
        return 3
    elif (x == 'twelve'):
        return 12
    else:
        return 0
coches['num-of-cylinders'] = coches['num-of-cylinders'].apply(text_to_number).astype('float')
```

```python
coches['num-of-doors'] = coches['num-of-doors'].apply (
    lambda x: 4 if x=='four' else 2
)
```

```python
num_cols = [
'symboling         ', 
' num-of-doors     ',  
' wheel-base       ',  
' length           ',  
' width            ',  
' height           ',  
' curb-weight      ',  
' num-of-cylinders ',  
' engine-size      ',  
' compression-ratio',  
' horsepower       ',  
' peak-rpm         ',  
' city-mpg         ',  
' highway-mpg      ',  
' price            ']
```

```python
numeric_columns = []

for col in num_cols: 
    numeric_columns.append(col.strip())
```

```python
numeric_columns
```

```python
coches_n = coches[numeric_columns].copy()
```

```python
coches_n = coches_n.astype('float')
```

```python
coches_n
```

```python
coches_n.isnull().sum()
```

```python
coches_n = coches_n.drop(columns = ['symboling'])
```

```python
coches_n.isnull().sum()
```

```python
coches_n = coches_n.dropna()
```

```python
coches_n.isnull().sum()
```

```python
price = coches_n['price']
coches_n = (coches_n - coches_n.mean())/coches_n.std()
coches_n['price'] = price
```

```python
coches_n
```

```python
from sklearn.neighbors import KNeighborsRegressor

```

```python
k_split = math.trunc(len(coches_n)/5)

for i in range(0,5):
    coches_n.loc[coches_n.index[(k_split*i):(k_split*(i+1) if i<4 else len(coches_n))], 'fold'] = i+1
```

```python
coches_n['fold'].value_counts()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def knn_train_test(train_cols, target_col, df, p = 2, k = 5, alg = 'uniform'):
    _df = df.copy()

    """
    Calcula la predicción y devuelve el rmse
    """
    split_value = math.trunc(len(_df)*0.8)
    train = _df.iloc[:split_value]
    test = _df.iloc[split_value:]
    
    knn = KNeighborsRegressor(n_neighbors = k, weights=alg, p=p)
    knn.fit(train[train_cols],train[target_col] )
    pred = knn.predict(test[train_cols])
    rmse = mean_squared_error(test[target_col], pred)**0.5
    
        
    return rmse
        
    
    
    
```

```python
knn_train_test(['num-of-doors', 'horsepower'], 'price', coches_n)
```

```python
coches_n.corr()
```

```python
knn_train_test(['engine-size', 'horsepower', 'highway-mpg'], 'price', coches_n)
```
