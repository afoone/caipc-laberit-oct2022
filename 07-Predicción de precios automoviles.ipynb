```python
def text_to_number(x):
    if (x == 'four'):
        return 4
    elif (x == 'six'):
        return 6
    elif (x == 'five'):
        return 5
    elif (x == 'eight'):
        return 8
    elif (x == 'two'):
        return 2
    elif (x == 'three'):
        return 3
    elif (x == 'twelve'):
        return 12
    else:
        return 0
coches['num-of-cylinders'] = coches['num-of-cylinders'].apply(text_to_number).astype('float')
```
