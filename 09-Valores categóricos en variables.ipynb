```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/afoone/caipc-laberit-oct2022/master/data/Ecom%20Expense.csv')
```

```python
# not to do 
df['Gender'].apply(lambda x: 1 if x=='Male' else 0)
```

```python
# not to do neither
df['City Tier'].apply(lambda x: 1 if x=='Tier 1' else (2 if x == "Tier 2" else 3))
```

```python
# variables dummies 
gender_dummies = pd.get_dummies(df['Gender'], prefix='gender')
gender_dummies
```

```python
city_dummies = pd.get_dummies(df['City Tier'], prefix='city')

city_dummies
```

```python
df = df.join(gender_dummies).join(city_dummies)
```
