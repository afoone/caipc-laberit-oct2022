```python
import pandas as pd 
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/afoone/caipc-laberit-oct2022/master/data/titanic.csv')
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
df
```

```python
dummies_embarked = pd.get_dummies(df['Embarked'], prefix='port')
```

```python
dummies_embarked
```

```python
gender_dummies  = pd.get_dummies(df['Sex'], prefix='gender')
```

```python
class_dummies = pd.get_dummies(df['Pclass'], prefix='class')
```

```python
df = df.join(gender_dummies).join(class_dummies).join(dummies_embarked)
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.dropna()
```

```python
X = df[['SibSp', 'Parch', 'Fare',
        'gender_female', 'gender_male', 'class_1', 'class_2',
       'class_3', 'port_C', 'port_Q', 'port_S']]
```

```python
y = df['Survived']
```

```python
# selección de variables
from sklearn import datasets
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
seleccion = RFE(lr, n_features_to_select=4, step=1).fit(X, y)
```

```python
seleccion.support_
```

```python
seleccion.ranking_
```

```python
list(zip(['SibSp', 'Parch', 'Fare',
        'gender_female', 'gender_male', 'class_1', 'class_2',
       'class_3', 'port_C', 'port_Q', 'port_S'], seleccion.ranking_))
```

```python
X=X[[
        'gender_female', 'gender_male', 'class_1',
       'class_3']]
```

```python
lr.fit(X, y)
```
